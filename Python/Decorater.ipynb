### Decorator

```
#decorator
def decorator(new_func):
  def sub_function():
    print("This is sub function inside main function")
    new_func()
    print("Block of code below")
  return sub_function()
```

```
@decorator
def new_feature():
  print("New lines of code replacing function call : new_func")
```
