### OOPS-Inheritance 

```
# all class variable are public
class Machine_learning ():
  def __init__(self,python,stats,data):
    self.python = python
    self.stats = stats
    self.data = data
  def prediction(self):
    print(f"One good ML model uses {self.stats} along with {self.python} on clean {self.data} to give prediction ")
```

```
ml = Machine_learning('python3','Central Limit theorem', 1)
ml.prediction()
```

```
class Deep_learning(Machine_learning):
  def __init__(self,python,stats,data,TPU):
    super().__init__(python,stats,data)
    self.TPU = TPU
  def resnet(self):
    print("Deep learning requires high computational power alongwith pre_requisite of ML model")
```

```
dl = Deep_learning('python3','CET', 1, 'Nvidia')
dl.resnet()
```

### Private Public Protected

```
# protected
# all class variable are protected
class Machine_learning_protected ():
  def __init__(self,python,stats,data):
    self._python = python
    self._stats = stats
    self._data = data
  def prediction(self):
    print("One good ML model uses statistics along with python on clean data to give prediction ")
```

```
ml = Machine_learning_protected('python3','Central Limit theorem', 1)
#ml.prediction()
dir(ml)
```

### Magic Method in Python

```
# class without__str__ method

class Machine_learning_without_str_method():
  def __init__(self,python,stats,data):
    self.python = python
    self.stats = stats
    self.data = data
  def prediction(self):
    print("One good ML model uses statistics along with python on clean data to give prediction ")
  def __sizeof__(self):
        return "This displays size of the object"
# class with__str__ method
class Machine_learning_with_str_method():
  def __init__(self,python,stats,data):
    self.python = python
    self.stats = stats
    self.data = data
  def prediction(self):
    print("One good ML model uses statistics along with python on clean data to give prediction ")
  def __str__(self):
        return "The object has been initialized ."
  def __sizeof__(self):
        return "This displays size of the object"
  def __repr__(self):
        return " string representation of an object that is a valid Python object"
```

```
ml1 = Machine_learning_without_str_method('PythonV3','Normal_distribution',1)
print(ml1)
ml2 = Machine_learning_with_str_method('PythonV3','Normal_distribution',1) 
print(ml2)
ml1.__sizeof__()
```

* summary: 
* str() & __str()__ return a printable string representation of an object... something human-readable/for human consumption
* repr() & __repr()__ return a string representation of an object that is a valid Python object, something you can pass to eval() or type into the Python shell without getting an error.
